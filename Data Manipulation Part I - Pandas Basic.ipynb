### Pandas loc and iloc
***

Panda use loc() and iloc() methods for manipulation data.

loc() is label based data selecting method which means that we have to pass the name of the row or column which we want to select.

iloc() is a indexed based selecting method which means that we have to pass integer index in the method to select specific row/column.

### Joining and Merging Pandas Data Frame
***

#### Join

```python
raw_data = {
    'subject_id': ['1','2','3','4'],
    'first_name': ['Alex','Allan', 'Adit','Frezco'],
    'last_name':['Anderson', 'Ackerman', 'Aoni','Ali']
}

df_a = pd.DataFrame(raw_data)

raw_data = {
  'subject_id': ['1','2','3','4'],
    'first_name': ['Alexa','Amy', 'Aadira','Freya'],
    'last_name':['Lopez', 'Horan', 'Styles','Twain']
}

df_b = pd.DataFrame(raw_data)
```

```python
df_a
```

```python
df_b
```

```python
#Join row
df_new = pd.concat([df_a, df_b])
```

```python
df_new
```

```python
#Join Column
df_new = pd.concat([df_a, df_b], axis=1)
df_new
```

#### Merge
***

```python
df1 = pd.DataFrame({'nama': ['foo', 'bar', 'baz', 'foo', 'bde'],
                    'height': [1, 2, 3, 5, 10]})
df2 = pd.DataFrame({'nama': ['foo', 'bar', 'baz', 'foo', 'asd'],
                    'depth': [5, 6, 7, 8, 9]})
```

```python
df2
```

```python
df1.merge(df2, left_on='nama', right_on='nama')
```
